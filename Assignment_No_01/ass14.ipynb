```python
import matplotlib.pyplot as plt

# Example data for used car prices in lakhs
used_car_prices = [5.2, 3.8, 7.1, 4.5, 6.0, 5.5, 8.2, 4.0, 6.5, 7.8, 5.9, 3.5, 4.8, 6.2, 7.0, 5.1, 6.8, 4.3, 5.7, 7.3]

plt.hist(used_car_prices, bins=20, color='Darkblue', edgecolor='black')
plt.title("Distribution of Used Car Prices")
plt.xlabel("Price in Lakhs")
plt.ylabel("Number of Cars")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

brands = ['Lamborghini', 'Bentley', 'Porsche', 'Jaguar', 'Land', 
          'Mini', 'Audi', 'Mercedes-Benz', 'BMW', 'Isuzu']
avg_prices = [4.8, 4.1, 3.7, 3.5, 3.3, 3.2, 3.1, 3.0, 3.0, 2.9] 
plt.bar(brands, avg_prices, color='darkblue', edgecolor='black')
plt.title("Brand Vs Price")
plt.xlabel("Brand")
plt.ylabel("Average Price (Lakhs)")
plt.xticks(rotation=45)

plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Example DataFrame for demonstration
data = pd.DataFrame({
	'fuel_type': ['Petrol', 'Diesel', 'Petrol', 'Diesel', 'Petrol', 'Diesel', 'Petrol', 'Diesel', 'Petrol', 'Diesel'],
	'price': [5.2, 3.8, 7.1, 4.5, 6.0, 5.5, 8.2, 4.0, 6.5, 7.8]
})

sns.barplot(x='fuel_type', y='price', data=data)
plt.title('Average Price by Fuel Type')
plt.xticks(rotation=45)
plt.show()
```

```python
data.hist(figsize=(12, 8), edgecolor='black')
plt.suptitle('Histograms of Numeric Features', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data)
plt.xticks(rotation=90)
plt.title("Boxplot of All Numeric Features")
plt.show()
```

```python
# Create a DataFrame for brands and avg_prices
brand_df = pd.DataFrame({'brand': brands, 'price': avg_prices})

sns.barplot(x='brand', y='price', data=brand_df)
plt.title("Average Price by Brand")
plt.xticks(rotation=45)
plt.show()
```

```python
# 'data' only contains 'price' and 'fuel_type', so we can only plot 'price'
sns.pairplot(data[['price']])
plt.suptitle("Pair Plot of Price Feature", y=1.02)
plt.show()
```
